```python
# Definir columnas numéricas
numericas = ["age", "gpa", "attendance_rate", "study_hours_per_week", "sleep_hours_per_night"]

# Calcular coeficiente de asimetría
asimetria = df[numericas].skew()
print("Coeficiente de asimetría:\n", asimetria)

# Crear gráficos de distribución
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i, col in enumerate(numericas):
    sns.histplot(df[col], bins=20, kde=True, ax=axes[i], color="royalblue", alpha=0.7)
    axes[i].set_title(f"Asimetría de {col}: {asimetria[col]:.2f}")

plt.tight_layout()
plt.show()
print(df.shape)
```

**Análisis del coeficiente de asimetría (skew)**  

El coeficiente de asimetría nos indica si los datos están distribuidos de manera simétrica o si presentan un sesgo hacia valores más altos o más bajos.  

### **Reglas generales**  
- **Si skew ≈ 0**, la distribución es simétrica.  
- **Si skew > 0**, los valores altos son más frecuentes (sesgo a la derecha).  
- **Si skew < 0**, los valores bajos son más frecuentes (sesgo a la izquierda).

**5.4 Relación entre horas de estudio y GPA**

```python
df[['study_hours_per_week', 'gpa']].dropna().head(10)

# Creación de intervalos de horas de estudio
bins = [0, 5, 10, 15, 20, 25, 30, 35, 40, 50]
labels = ['0-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35', '36-40', '41-50']
df['study_hours_group'] = pd.cut(df['study_hours_per_week'], bins=bins, labels=labels, right=False)

# Calculo GPA promedio por grupo
gpa_promedio = df.groupby('study_hours_group')['gpa'].mean().reset_index()
print(gpa_promedio)

# Gráfico de barras
sns.barplot(x='study_hours_group', y='gpa', data=gpa_promedio)
plt.title('GPA promedio según rangos de horas de estudio por semana')
plt.xlabel('Horas de estudio por semana')
plt.ylabel('GPA promedio')
plt.show()
```

**Análisis de la relación entre horas de estudio y GPA promedio**


* El análisis muestra cómo el GPA varía según la cantidad de horas dedicadas al estudio semanal. Se observan las siguientes tendencias:

* El GPA más alto se encuentra en el grupo que estudia entre 21 y 25 horas semanales (2.16), lo que sugiere que un tiempo de estudio moderado-alto se asocia con mejor rendimiento académico.

* Los GPA más bajos aparecen en los grupos de 6-10 horas (1.80) y 41-50 horas (1.67), indicando que tanto estudiar muy poco como en exceso podría no ser óptimo.

No hay una relación lineal directa entre más horas de estudio y mejor desempeño académico. Algunos grupos con mayor carga horaria no presentan un GPA más alto (por ejemplo, 31-35 horas tiene 1.75, mientras que 36-40 horas sube a 1.99).

**Interpretación**

Estos resultados sugieren que la cantidad de horas no es el único factor determinante en el rendimiento académico. Otros elementos, como la calidad del estudio, técnicas de aprendizaje y el equilibrio con descanso y actividades extracurriculares, pueden influir significativamente.

**Conclusión**

Existe un rango óptimo de estudio (21-25 horas semanales) asociado con un mejor desempeño académico. Estudiar demasiado poco o en exceso podría no ser efectivo, destacando la importancia de un estudio planificado y balanceado.
